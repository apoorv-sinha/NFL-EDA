```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
%matplotlib inline
```

```python
WR = pd.read_csv('WR.csv', engine='python')
```

```python
WR.head()
```

```python
phi = np.linspace(0, 2*np.pi, 21)
x = np.sin(phi)
y = np.cos(phi)
rgb_cycle = np.vstack((            # Three sinusoids
    .5*(1.+np.cos(phi          )), # scaled to [0,1]
    .5*(1.+np.cos(phi+2*np.pi/3)), # 120Â° phase shifted.
    .5*(1.+np.cos(phi-2*np.pi/3)))).T # Shape = (60,3)
```

```python
plt.figure(figsize = (10, 10))
sns.scatterplot(data = WR, x = 'Receiving_TDs', y = 'Rushing_TDs', c = rgb_cycle, s = 90)

plt.title("FINDING UNICORNS AMONG WRs")
plt.xlabel("Receiving TDs")
plt.ylabel("Rushing TDs")

for i in range(WR.shape[0]):
    plt.text(x = WR.Receiving_TDs[i]+0.3,y= WR.Rushing_TDs[i]+0.3,s= WR.Wide_Receiver_Name[i], 
          fontdict=dict(color= 'blue',size=10),
          bbox=dict(facecolor='white',alpha=0.5))
```
