```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
%matplotlib inline
```

```python
QB = pd.read_csv('QB.csv', engine = 'python')
```

```python
QB.head()
```

```python
TDr = QB['Rushing_TD']
TDp = QB['Passing_TD']
```

```python
name = QB['QB_name']
lables = ['Rushing TDs', 'Passing TDs']
L = []
for i in range(QB.shape[0]):
    plt.title(name[i])
    L.append(TDr[i])
    L.append(TDp[i])
    plt.pie(L, labels = lables, autopct = "%1.1f%%", shadow = True, startangle = 140)
    plt.show()
    L.clear()
```

```python
RYrd = QB['Rushing_Yards']
PYrd = QB['Passing_Yards']
```
