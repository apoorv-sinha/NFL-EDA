```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
phi = np.linspace(0, 2*np.pi, 19)
x = np.sin(phi)
y = np.cos(phi)
rgb_cycle = np.vstack((            # Three sinusoids
    .5*(1.+np.cos(phi          )), # scaled to [0,1]
    .5*(1.+np.cos(phi+2*np.pi/3)), # 120° phase shifted.
    .5*(1.+np.cos(phi-2*np.pi/3)))).T # Shape = (60,3)
```

```python

fig = plt.figure(figsize = (10, 10))
ax = fig.add_subplot(projection='3d')

z = Def['Fumbles_Forced'].values
x = Def['Tackles'].values
y = Def['Sacks'].values
w = Def['Player_Name'].values

ax.scatter(x, y, z, c= rgb_cycle, marker='o')



plt.title("FINDING UNICORNS AMONG FIRST LINE OF DEFENSE")
plt.xlabel("Tackles")
plt.ylabel("Sacks")
ax.set_zlabel("Fumbles Forced")

for x,y,z,i in zip(x,y,z,w):
    ax.text(x,y,z,i)
                     
            
for angle in range(0, 360):
    ax.view_init(30, angle)
    plt.draw()
    plt.pause(.001)
```
